```python
from google.cloud import bigquery  # For interacting with Google BigQuery
from google.oauth2 import service_account  # For service account authentication
from google.auth.transport.requests import Request  # For making authenticated HTTP requests
import google.auth  # For handling authentication
from googleapiclient.discovery import build  # For building and interacting with Google APIs
import pandas as pd # For the dataset
```

```python
KEY_PATH = r"C:\Users\****\Downloads\atlantean-site-430006-t5-86942fca642e.json"

credentials = service_account.Credentials.from_service_account_file(KEY_PATH) # Authenticate with the credentials
```

```python
# Initialize the BigQuery client
client = bigquery.Client(credentials=credentials, project=credentials.project_id)
```

```python
dataset_id = 'atlantean-site-430006-t5.product_analytics'

# Get the dataset reference
dataset_ref = client.get_dataset(dataset_id)

# List all tables in the dataset
tables = client.list_tables(dataset_ref)
for table in tables:
    print(f"Table name: {table.table_id}")   
```

```python
query = """
SELECT *
FROM `atlantean-site-430006-t5.product_analytics.sachin ODI`
LIMIT 10
"""

# Run the query
query_job = client.query(query)

results = query_job.result()
```
